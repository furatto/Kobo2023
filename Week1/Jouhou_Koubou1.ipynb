```python
import numpy as np

#課題1 AND NAND OR XOR
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    y = np.sum(w*x) + b
    if y <= 0:
        return 0
    else:
        return 1

def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array ([-0.5,-0.5])
    b = 0.7
    y = np.sum(w*x) + b

    if y <= 0:
        return 0
    else:
        return 1
        
    
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.4
    y = np.sum(w*x) + b

    if y <= 0:
        return 0
    else:
        return 1
    
def XOR(x1, x2):  
    if NAND(x1,x2) == 1 and OR(x1,x2) == 1:
        return 1
    else:
        return 0

#課題2 真理値出力確認
if __name__ == '__main__':
    
    print("AND")
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        #AND
        out = AND(xs[0], xs[1])
        print(str(xs) + " -> " + str(out))

    print("NAND")
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        #NAND
        out = NAND(xs[0],xs[1])
        print(str(xs) + " -> " + str(out))
    
    print("OR")
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        #OR
        out = OR(xs[0],xs[1])
        print(str(xs) + " -> " + str(out))

    print("XOR")
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        #XOR
        out = XOR(xs[0],xs[1])
        print(str(xs) + " -> " + str(out))

#課題4 Sigmoid Softmax関数
def Sigmoid(x):
    h = np.empty(0)
    for i in x:
        y = 1/(1+ np.exp(-x[i-1]))
        h = np.append(h,y)
    return h

def Softmax(x):
    h = np.empty(x.size)
    
    y_down = 0
    for i in x:
         y_down += np.exp(x[i-1])
    for i in x:
        y_up = np.exp(x[i-1])
        y = y_up/y_down
        
        h[i-1] = y
    return h

def init_network():
    network={}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([[0.1, 0.2, 0.3]])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([[0.1, 0.2]])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([[0.1, 0.2]])
    return network
    
def forward(network, x):
    # layer 1
    W1=network['W1']
    b1=network['b1']
    a1=np.dot(x,W1)+b1
    z1=Sigmoid(a1)
    
    # layer2
    W2 = network["W2"]
    b2 = network["b2"]
    a2 = np.dot(z1,W2) + b2
    z2 = Sigmoid(a2)
    
    #layer3
    W3 = network["W3"]
    b3 = network["b3"]
    a3 = np.dot(z2,W3) + b3
    y = Softmax(a3)
    
    return y

network = init_network()
x=np.array([1.0,0.5])
y=forward(network,x)
print(y) #[0.40625907, 0.59374093]

#課題5 バッチ処理の行い方
network = init_network()
batch_size = 16 # batchサイズ

x=np.random.rand(1000,2)
for i in range(0,len(x),batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = forward(network,x_batch)

print(y_batch)
    
```
